```python
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
# VQC

plt.style.use("seaborn")
fig = plt.figure(figsize=(12, 8))

samples, wine_time, wine_acc, iris_time, iris_acc = np.loadtxt('vqc.csv', delimiter = ",", unpack=True)

plt.plot(samples, wine_time, marker='o',color='purple', linestyle=':', markersize=10, label='wine dataset')
plt.plot(samples, iris_time, marker='o',color='blue', linestyle=':', markersize=10, label='iris dataset')

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.xlabel('Number of Samples', fontsize=20)
plt.ylabel('Time Taken (in seconds)', fontsize=20)
plt.title('Runtime vs. Number of Samples for Wine and Iris Datasets', fontsize=20)
plt.legend(fontsize=20)
#plt.yscale('log')

plt.show()
```

```python
# QGAN

plt.style.use("seaborn")
fig = plt.figure(figsize=(12, 8))

size, avg_time, min_time, max_time, avg_acc, min_acc, max_acc = np.loadtxt('qgan.csv', delimiter = ",", unpack=True)

plt.plot(size, avg_time, marker='o',color='purple', linestyle=':', markersize=10)
plt.fill_between(size, min_time, max_time, facecolor='purple', alpha = 0.2)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xscale('log')

plt.xlabel('Array Size of Weights', fontsize=20)
plt.ylabel('Time Taken (in seconds)', fontsize=20)
plt.title('Runtime vs Array Size of QGAN Weights', fontsize=20)
plt.legend(fontsize=20)
#plt.yscale('log')

plt.show()
```
