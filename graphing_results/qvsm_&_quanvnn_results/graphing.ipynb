# GRAPH QAUNV RESULTS

```python
import matplotlib.pyplot as plt
import csv
import numpy as np
import pandas as pd
```

```python
import csv
csvfile = open('quanvnn.csv','r')
csvFileArray = []

for row in csv.reader(csvfile, delimiter = ','):
    for term in row:
        csvFileArray.append(float(term))
    #print(term)

#print(csvFileArray)

#samples = csvFileArray[0]

samples_float = csvFileArray[0:6]
q_acc = csvFileArray[6:12]
c_acc = csvFileArray[12:18]
q_time = csvFileArray[18:24]
c_time = csvFileArray[24:30]

print('\n',samples_float)
print(q_acc)
print(c_acc)
print(q_time)
print(c_time)




samples = []
for i in samples_float:
    samples.append(int(i))
print(samples)
```

```python
plt.style.use("seaborn")
fig = plt.figure(figsize=(12, 8))
plt.plot(samples, q_acc, marker='o',label='Quantum Layers',  linestyle=':', color='blue', markersize=10)

plt.plot(samples, c_acc, marker='o',label='Classical Layers',  linestyle=':', color='purple', markersize=10)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.xlabel('Number of Samples', fontsize=20)
plt.ylabel('Accuracy', fontsize=20)
plt.title('Accuracy vs. Number of Samples (30 Epochs)', fontsize=20)
plt.legend(fontsize=20)

plt.show()
plt.tight_layout()

fig = plt.figure(figsize=(12, 8))

plt.plot(samples, q_time, marker='o',label='Quantum Layers',  linestyle=':', color='blue', markersize=10)


plt.plot(samples, c_time, marker='o',label='Classical Layers',  linestyle=':', color='purple', markersize=10)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.xlabel('Number of Samples', fontsize=20)
plt.ylabel('Run-time (in s)', fontsize=20)
plt.title('Run-time vs. Number of Samples (30 Epochs)', fontsize=20)
plt.legend(fontsize=20)

#plt.yscale()
```
