```python
import matplotlib.pyplot as plt
import csv
import numpy as np
import pandas as pd

```

```python

import csv
csvfile = open('qaoa-depth1.csv','r')
csvFileArray = []
for row in csv.reader(csvfile, delimiter = ','):
    for term in row:
        csvFileArray.append(float(term))

    
n = csvFileArray[0:5]
nodes = []
for i in n:
    nodes.append(int(i))

all_to_all = csvFileArray[5:10]
nearest_n = csvFileArray[10:15]
random = csvFileArray[15:20]
#print(csvFileArray)
#print('\n',nodes)
```

```python
plt.style.use("seaborn")
fig = plt.figure(figsize=(12, 8))

plt.plot(nodes, all_to_all, marker='o',label='All-to-all ', color='purple', linestyle=':', markersize=10)
plt.plot(nodes, nearest_n, marker='o',label='Nearest Neighbour',color='blue',  linestyle=':', markersize=10)
plt.plot(nodes, random, marker='o',label='Random',  linestyle=':', color='red', markersize=10)


#plt.plot(nodes, all_to_all_2, marker='o',label='All-to-all 2 ', color='green', linestyle=':', markersize=10)
#plt.plot(nodes, nearest_n_2, marker='o',label='Nearest Neighbour 2',color='orange',  linestyle=':', markersize=10)
#plt.plot(nodes, random_2, marker='o',label='Random 2',  linestyle=':', color='pink', markersize=10)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.xlabel('Number of Nodes', fontsize=20)
plt.ylabel('Time Taken (in seconds)', fontsize=20)
plt.title('Number of Nodes vs. Time. p (depth) = 1', fontsize=20)
plt.legend(fontsize=20)
#plt.yscale()
plt.show()
```

```python
csvfile = open('vqe.csv','r')
csvFileArray3 = []
for row in csv.reader(csvfile, delimiter = ','):
    for term in row:
        csvFileArray3.append(float(term))

qubits = csvFileArray3[0:3]
runtime_lih = csvFileArray3[3:6]
print(runtime_lih)
```
