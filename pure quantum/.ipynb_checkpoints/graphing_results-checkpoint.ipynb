```python
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
# GROVERS ALGORITHM

plt.style.use("seaborn")
fig = plt.figure(figsize=(12, 8))

qubits, runtime = np.loadtxt('grovers.csv', delimiter = ",", unpack=True)

plt.plot(qubits, runtime, marker='o',color='blue', linestyle=':', markersize=10)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.xlabel('Number of Qubits', fontsize=20)
plt.ylabel('Time Taken (in seconds)', fontsize=20)
plt.title('Qubits vs. Time. Grovers Algorithm', fontsize=20)
plt.legend(fontsize=20)
#plt.yscale('log')

plt.show()
```
