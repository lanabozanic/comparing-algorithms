```python
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
# QUANTUM PHASE ESTIMATION

plt.style.use("seaborn")
fig = plt.figure(figsize=(12, 8))

qubits, runtime, minimum, maximum = np.loadtxt('qpe-18.csv', delimiter = ",", unpack=True)

plt.plot(qubits, runtime, marker='o',color='orange', linestyle=':', markersize=10)
plt.fill_between(qubits, minimum, maximum, facecolor='orange', alpha = 0.1)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.xlabel('Number of Qubits', fontsize=20)
plt.ylabel('Time Taken (in seconds)', fontsize=20)
plt.title('Qubits vs. Time for Quantum Phase Estimation (Up to 18 Qubits)', fontsize=20)
plt.legend(fontsize=20)
#plt.yscale('log')

plt.show()
```
